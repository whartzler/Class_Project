```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
file = '../Brand_mapping/FCV_Dataset_mapped.csv'
df = pd.read_csv(file)

df.head()
```

```python
df = df.drop(columns=['Unnamed: 0', 'submitdate. Date submitted', 'Month/Year Submitted', 'Month Submitted', 'Year Submitted', 'lastpage. Last page'])
df.head()
```

```python
le = LabelEncoder()
```

```python
df['Manufacture'] = le.fit_transform(df['Manufacture'])
df['Model'] = le.fit_transform(df['Model'])

df.head()
```

```python
df.to_csv('FCV_Label_Encoded.csv', index=False)
```
