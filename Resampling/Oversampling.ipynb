```python
import pandas as pd
from sklearn.utils import resample

file = '../Encoding/FCV_Label_Encoded.csv'
df = pd.read_csv(file)

df
```

```python
df = df.rename(columns={'FCV, BEV Dummy' : 'Target'})
```

```python
df_majority = df[df.Target==1.0]
df_minority = df[df.Target==0.0]
```

```python
df_minority_oversampled = resample(df_minority,
                                   replace=True,
                                   n_samples=len(df_majority),
                                   random_state=123)
```

```python
df_oversampled = pd.concat([df_majority, df_minority_oversampled])
```

```python
df_oversampled['Target'].value_counts()
```

```python
df_oversampled.to_csv('FCV_Oversampled.csv', index=False)
```
